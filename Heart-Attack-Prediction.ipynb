### 1.1 Overview of Project Topic <a id = 2></a>

### What is a heart attack?

A heart attack (myocardial infarction) is the occlusion of a heart vessel by plaque-like lesions filled with cholesterol and fat. Due to this blockage, there is a cut-off of blood flow towards the heart which can lead to death.


### Causes of a Heart Attack

<ul>
    <li><strong>Coronary artery disease </strong> is the leading cause of heart attacks</li>
    <li><strong>Coronary artery spasm: </strong> severe squeezing of unblocked blood vessel </li>
    <li><strong>Infections: </strong> COVID-19 as well as other viral infections damage the heart muscle </li>
    <li><strong>Spontaneous coronary artery dissection (SCAD): </strong> internal heart artery tear </li>
</ul>

### Risk Factors

<ul>
    <li>High Blood Pressure</li>
    <li>Obesity</li>
    <li>High cholesterol</li>
    <li>Diabetes</li>
    <li>Age</li>
    <li>Stress</li>
    <li>Lack of exercise</li>
    <li>Genetic prevalence</li>
    <li>Unhealty diet</li>
</ul>

### 1.2 Dataset Variables <a id = 3></a>

- **_Age_**: Patient age
- **_Sex_**: Patient sex
- **_exang_**: Exercise Induced Angina (0 = No; 1 = Yes)
- **_ca_**: Number of major blood vessels
- **_cp_**: Type of chest pain
- **_trtbps_**: Patient resting blood pressure (mmHg)
- **_chol_**: Patient cholestorol level (mm/dL)
- **_fbs_**: Patient fasting blood sugar (> 120 mg/dL) (0 = False; 1 = True)
- **_rest_ecg_**: Patient resting electrocardiograph results
- **_thalach_**: Maximum recorded heart rate
- **_target_**: Chance of heart attack (0 = Lesser; 1 = Greater)

**Additional Variable Descriptions**
1. age - age in years

2. sex - sex (1 = male; 0 = female)

3. cp - chest pain type (1 = typical angina; 2 = atypical angina; 3 = non-anginal pain; 0 = asymptomatic)

4. trestbps - resting blood pressure (in mm Hg on admission to the hospital)

5. chol - serum cholestoral in mg/dl

6. fbs - fasting blood sugar > 120 mg/dl (1 = true; 0 = false)

7. restecg - resting electrocardiographic results (1 = normal; 2 = having ST-T wave abnormality; 0 = hypertrophy)

8. thalach - maximum heart rate achieved

9. exang - exercise induced angina (1 = yes; 0 = no)

10. oldpeak - ST depression induced by exercise relative to rest

11. slope - the slope of the peak exercise ST segment (2 = upsloping; 1 = flat; 0 = downsloping)

12. ca - number of major vessels (0-3) colored by flourosopy

13. thal - 2 = normal; 1 = fixed defect; 3 = reversable defect

14. num - the predicted attribute - diagnosis of heart disease (angiographic disease status) (Value 0 = < diameter narrowing; Value 1 = > 50% diameter narrowing)
